```python
# wrap method chain in a function
path = 'data/raw/waterpotability.csv'

def load_and_process(path):
    df = (
        pd.read_csv(path)
        .rename(columns={'ph': 'pH',
              'Hardness': 'Hardness (mg/L)',
              'Solids': 'Solids (ppm)',
              'Chloramines': 'Chloramines (ppm)',
              'Sulfate': 'Sulfate (mg/L)',
              'Conductivity': 'Conductivity (μS/cm)',
              'Organic_carbon': 'Organic Carbon (ppm)',
              'Trihalomethanes': 'Trihalomethanes (μg/L)',
              'Turbidity': 'Turbidity (NTU)',
              'Potability': 'Potable'})
        .dropna().reset_index(drop=True)
        .replace(1, 'Yes')
        .replace(0, 'No')
     )
    
    return df
```

```python
df1 = (df
      .groupby('Potable')['pH']
      .mean()
      .sort_values(ascending=True)
      .reset_index()
      )
      
df1
```

```python
df2 = (df
      .groupby('Potable')['Hardness (mg/L)']
      .mean()
      .sort_values(ascending=True)
      .reset_index()
      )
df2
```

```python
df['Chemicals'] = df['Chloramines (ppm)'] +\
                       df['Sulfate (mg/L)'] +\
                       df['Organic Carbon (ppm)'] +\
                       df['Trihalomethanes (μg/L)']
df
```

```python
chemicals_df = (df
            .groupby('Potable')['Chemicals']
            .mean()
            .sort_values(ascending=True)
            .reset_index()
            )
chemicals_df
```

```python
# the difference between non-potable water and potable water is very small when looking at the data one column at a time

import matplotlib.pyplot as plt
plt.figure(figsize=(6,6))
sns.violinplot(data=df, x='Potable', y='Chemicals', color='purple')
plt.title('Potability Depending on Amount of Chemicals')
plt.ylabel('')
plt.xlabel('Potable')
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(data=df, x='pH', y='Potable', color='green', order=df1['Potable'].tolist(), hue="Potable")
plt.title('Box plot of Potability determined by pH')
plt.ylabel('')
plt.xlabel('pH of Water')
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(data=df, x='Hardness (mg/L)', y='Potable', color='green', order=df2['Potable'].tolist(), hue='Potable')
plt.title('Box plot of Potability determined by Hardness')
plt.ylabel('')
plt.xlabel('Hardness of Water')
# the medians of the hardness of water are both very close for potable and non-potable water
```

# Task 4 - Conduct Analysis


### What factor(s) is most important in determining whether drinking water is potable or non-potable?

Clean and safe drinking water is an essential need for human beings. The description of the dataset even states that in some cases, investing in clean drinking water can benefit the economy as the health effects caused by non-potable water would decrease. I initially wanted to research more about different types of bodies of water and whether they were potable or not. However, this dataset does not specify any names of the bodies of water and instead gives parameters to help determine the potability. The parameters used in this dataset are pH, hardness, total dissolved solids, chloramines, sulphate, conductivity, total organic carbon, trihalomethanes, and turbidity. Using this information, I decided to base my analysis around what factor is most important in determining whether drinking water is potable or non-potable. 

I decided to first investigate how much the pH affected the potability in this dataset. The pH for pure water is 7, but the dataset states that according to WHO standards, the acceptable range is from 6.5 to 8.5. According to my analysis, both calculated mean values of the pH for potable and non-potable are within these standards. The boxplot shows that for both potable and non-potable, the majority of the values lie within the acceptable range. The plot also shows that even when the pH is not within that range, it is still potable. I also looked at the hardness of the water. These values, as seen in the boxplot, are also very similar. The medians are just about the same and they also have almost the same range. This indicates that the hardness of water is probably not the most important factor. To examine more columns at once I combined the parameters that I believe are chemicals. This also to my surprise has about the same median for both potable and non-potable. The shape also shows that the non-potable has more values concentrated around the median than the potable values. Lastly, a histogram plot shows that the count for non-potable water is greater than the count for potable drinking water. 


From my analysis, I can conclude that there is no one most important factor which determines whether water is potable. This can be seen from how the results for non-potable and potable look so similar when looking at pH, hardness, and chemicals.  All these parameters may greatly affect the potability of water, which helps establish why it is so important to invest in safe drinking water. It is very interesting to see how many parameters of water influence potability.
